```python
# Your code here

#avg_mental_health = df.groupby('HadDiabetes')['MentalHealthDays'].me

# Plot
plt.figure(figsize=(15, 5))
sns.barplot(x='HadDiabetes', y='MentalHealthDays', data=df1)
plt.title(' Diabetics and Non-diabetics')
plt.xlabel('Diabetes')
plt.ylabel(' Mental Health Days')
plt.show()
```

```python
df2 = df.copy()
req_columns = ['Sex', 'HadDepressiveDisorder' ]
df2 = df2[req_columns]
df2.isnull().sum()
df2 = df2[['Sex', 'HadDepressiveDisorder']].dropna()
df2.nunique()
```
