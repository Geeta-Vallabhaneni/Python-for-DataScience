### Task 2
**Question:** Are smokers heavier on average?

```python
# Your code here
df2= df.copy()
req_columns = ['SmokerStatus', 'WeightInKilograms' ]
df2 = df2[req_columns]
df2

```

```python
df_clean = df2[['SmokerStatus', 'WeightInKilograms']].dropna()
df_clean
#df_clean.nunique()
```

```python
Never_smoker = df_clean[ df_clean['SmokerStatus']=='Never smoked' ]
Former_smoker = df_clean[ df_clean['SmokerStatus']=='Former smoker' ]
Rare_smoker = df_clean[ df_clean['SmokerStatus']=='Current smoker - now smokes some days' ]
Daily_smoker = df_clean[ df_clean['SmokerStatus']=='Current smoker - now smokes every day' ]
Never_smoker
Former_smoker
Rare_smoker
Daily_smoker
```

```python

Never_smoker_avg=Never_smoker['WeightInKilograms'].mean()
Former_smoker_avg=Former_smoker['WeightInKilograms'].mean()
Rare_smoker_avg=Rare_smoker['WeightInKilograms'].mean()
Daily_smoker_avg=Daily_smoker['WeightInKilograms'].mean()
Never_smoker_avg

    
```

```python
plt.figure(figsize=(8, 6))
smoker_avg_weights = {
    'Never': Never_smoker_avg,
    'Former': Former_smoker_avg,
    'Rare': Rare_smoker_avg,
    'Daily': Daily_smoker_avg
}

# Plot

plt.bar(smoker_avg_weights.keys(), smoker_avg_weights.values())
plt.title('Average Weight by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Average Weight (kg)')

plt.show()
```
