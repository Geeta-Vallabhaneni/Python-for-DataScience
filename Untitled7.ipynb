### 11. Categorize students based on average into Grade column
> Add a column `Grade`:

* "A" if avg >= 90
* "B" if avg >= 75
* "C" if avg >= 60
* "D" otherwise


```python

def grade(avg):
    if avg >= 90:
        return 'A'
    elif avg >= 75:
        return 'B'
    elif avg >= 60:
        return 'C'
    else:
        return 'D'
grades=result.copy()
grades['Grade'] = grades['Average_Marks'].apply(grade)
grades[['Name', 'Average_Marks', 'Grade']]
```

```python
### 12. Count number of students in each grade
grade_count=grades.copy()
grade_count = grade_count.groupby('Grade').size().reset_index(name='Count')

print(grade_count)
```
